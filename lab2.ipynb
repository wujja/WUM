```python
import numpy as np
```

```python
def my_add(x, y):
  assert len(x.shape) == 2
  assert x.shape == y.shape
  x = x.copy()
  for i in range(x.shape[0]):
    for j in range(x.shape[1]):
      x[i,j] += y[i,j]
  return x  
```

```python
x = np.random.random((300,20))
y = np.random.random((300,20))
x, y, x.shape, y.shape, my_add(x, y)
```

```python
import time
start = time.time()

for i in range(10000):
  z = my_add(x, y)

print(f"Upłyneło: {time.time() - start} czasu")
```

```python
start = time.time()
for i in range(10000):
  z = x + y

print(f"Upłyneło: {time.time() - start} czasu")
```
